```python
import numpy as np
import plotly.graph_objects as go

# Función para calcular el orbital 1s del átomo de hidrógeno
def orbital_1s(r):
    return (1/np.sqrt(np.pi)) * np.exp(-r)

# Crear una malla tridimensional de puntos
r = np.linspace(0, 10, 100)
theta = np.linspace(0, np.pi, 100)
phi = np.linspace(0, 2*np.pi, 100)
r, theta, phi = np.meshgrid(r, theta, phi)

# Convertir las coordenadas esféricas a cartesianas
x = r * np.sin(theta) * np.cos(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(theta)

# Calcular los valores de la función orbital en cada punto de la malla
psi = orbital_1s(r)

# Crear la figura interactiva
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, surfacecolor=psi)])

# Ajustar el diseño de la gráfica
fig.update_layout(scene=dict(
                    xaxis_title='X',
                    yaxis_title='Y',
                    zaxis_title='Z'),
                  title='Orbital 1s del átomo de hidrógeno')

```
